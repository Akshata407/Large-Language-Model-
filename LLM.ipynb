```python
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import nltk

nltk.download('punkt', force=True) # Re-download the punkt tokenizer data to ensure integrity
nltk.download('punkt_tab') # Download the specific 'punkt_tab' resource as suggested by the error

# Define corpus
corpus = [
    "large language models are transforming natural language processing",
    "word2vec learns word embeddings from context",
    "natural language understanding is important"
]

print("word2vec")

# Tokenize sentences
sentences = [
    word_tokenize(sentence.lower())
    for sentence in corpus
]

# Train Word2Vec model
w2v_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=1
)

print("Vocabulary:", list(w2v_model.wv.key_to_index.keys()))
print("Vector size:", w2v_model.wv.vector_size)
```

```python
print("Word2Vec embedding(first 10 values) for 'language':")
print(w2v_model.wv['language'][:10], "...")
```
